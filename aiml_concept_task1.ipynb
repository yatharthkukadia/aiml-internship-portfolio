```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Iris.csv')
```

```python
data.head()
```

```python
plt.figure(figsize=(5,5))
sns.pairplot(data=data,hue='Species')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
# SepalLength
x = data['SepalLengthCm'].to_numpy()
```

```python
sl_mean = np.mean(x)
sl_median = np.median(x)
sl_std = np.std(x)
```

```python
print(f'SepalLength Mean: {sl_mean}, SepalLength Median: {sl_median}, SepalLength Standard Deviation: {sl_std}')
```
